```python
import numpy as np
import pandas as pd
import cv2 as cv
from google.colab.patches import cv2_imshow
from skimage import io
import matplotlib.pylab as plt

# Daftar URL gambar
urls = [
    "https://iiif.lib.ncsu.edu/iiif/0052574/full/800,/0/default.jpg",
    "https://iiif.lib.ncsu.edu/iiif/0016007/full/800,/0/default.jpg",
    "https://placekitten.com/800/571"
]

for url in urls:
    # Baca gambar dari URL
    image = io.imread(url)

    # Resize gambar (50%)
    image = cv.resize(image, (0, 0), fx=0.5, fy=0.5)

    # Konversi ke RGB
    image_rgb = cv.cvtColor(image, cv.COLOR_BGR2RGB)

    # Duplikasi gambar untuk dibandingkan
    image_rgb_copy = image_rgb.copy()

    # Gabungkan gambar sebelum-sesudah konversi
    final_frame = cv.hconcat([image, image_rgb])
    cv2_imshow(final_frame)
    print("\n")

    # Ambil resolusi
    tinggi, lebar = image_rgb.shape[:2]
    print("Resolusi image: tinggi x lebar =", tinggi, "x", lebar)

    # Tampilkan gambar asli (RGB)
    cv2_imshow(image_rgb)

    # Buat salinan untuk digaris
    image_with_line = image_rgb.copy()

    # Gambar garis horizontal di tengah
    for y in range(lebar):
        image_with_line[int(tinggi / 2), y] = [255, 255, 255]

    # Gabungkan gambar asli dan bergaris
    final_frame = cv.hconcat([image_rgb, image_with_line])
    cv2_imshow(final_frame)
    print("=" * 50)
```
